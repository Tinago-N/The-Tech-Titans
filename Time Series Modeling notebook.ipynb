### 1. Readind in the data

```python
data = pd.read_excel('N:/15.11.2023 Data/dataexport_20231115T124544.xlsx')
```

```python
data.head()
```

### 2. Data Cleaning

First 4 rows(3 plus current heading) are the same thing which can be replaced

The data is from Basel(location) found at latitude=47.75 and longitude=7.5. Resolution is measured hourly with no aggregation

```python
data = data.iloc[4:]
```

```python
data = data.drop([5,7])
data.head()
```

```python
data.columns = data.iloc[2]
data = data.drop(8)
data = data.drop(6)

# After dropping the necessary rows, now we can rename the columns accordingly with their unit of measurent
# Drop the 'Basel ' prefix from all column names
data.columns = data.columns.str.replace('Basel ', '')

# Concatenate column names with row 1 values in brackets
data.columns = [f'{col} ({data.iloc[0][col]})' for col in data.columns]

# Drop the first row, as it's now included in the column names
data = data.iloc[1:]

# Now we reset the index of the dataframe
data.reset_index()

data.head()
```

```python
data.columns
```

```python
# Rename the first column of the dataset from 'Timestamp Unit' to just 'Timestamp'
data = data.rename(columns={"timestamp (unit)": "Timestamp"})
```

```python
data.tail()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# 209280-191 = 209089
print(data.iloc[209088])
```

This suggests that the last 191 tuples are just null valued, so we drop the last 191 rows

```python
data = data.iloc[:-191]
data.tail()
```

```python
data.isnull().sum()
```

Now that there are no more null values, we then have to seperate the Timestamp column into data and time

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data['Date'], data['Time'] = data['Timestamp'].dt.normalize(), data['Timestamp'].dt.time
data = data.drop('Timestamp', axis=1)

# Move the two new columns to the begining of the dataframe
data = data[['Date', 'Time'] + [col for col in data.columns if col not in ['Date', 'Time']]]

data.head()
```
